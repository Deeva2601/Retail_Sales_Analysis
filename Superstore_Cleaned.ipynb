```python
import pandas as pd
```

```python
df = pd.read_csv(r"/content/Raw_Superstore_Data.csv")
df.head(5)
```

```python
df.columns
df.info()
df.describe()
```

```python
df.isnull().sum()
```

## Remove the NULL value in Sales,Quantity and discount because in numeric data we use mean ,median

```python
df['Sales'].fillna(df['Sales'].median(),inplace = True)
df['Quantity'].fillna(df['Quantity'].mean(),inplace = True)
df['Discount'].fillna(df['Discount'].mean(),inplace = True)
```

```python
df.head(7)
```

```python
df['Segment'].fillna(df['Segment'].mode())
```

```python
df.isnull().sum()
```

# we remove the all category nulls using mode

```python
categorial_cols = ['Customer Name','Segment','Region','City','Product Category','Product Name']
for cols in categorial_cols:
  df[cols].fillna(df[cols].mode()[0],inplace = True)

```

```python
categorical_null_counts = df[categorial_cols].isnull().sum()
```
