```python
import pandas as pd
```

```python
df = pd.read_csv("/content/Superstore_Cleaned (2).csv")
```

```python
df
```

```python
df['Quantity'] = df['Quantity'].astype(int)
```

```python
df['Discount'] = df['Discount'].round(2)
```

```python
df.head(8)
```

```python
df.shape
df.columns
df.describe()
```

```python
df.reset_index(drop=True, inplace=True)
```

Total Sales

```python
total_sales = df['Sales'].sum()
print("Total_sales: ",total_sales)
```

Category-wise Sales

```python
Category_sales = df.groupby('Product Category')['Sales'].sum().sort_values(ascending=False)
print(Category_sales)
```

```python
Category_sales = df.groupby('Product Category')['Sales'].sum().sort_values(ascending=False)
Category_sales.plot(kind='bar',title='Sales by category')
```

Product-wise Sale

```python
Product_Sales = df.groupby('Product Name')['Sales'].count().sort_values(ascending=False)
print(Product_Sales)
```

```python
Product_Sales = df.groupby('Product Name')['Sales'].count().sort_values(ascending=False)
Product_Sales.plot(kind = 'pie',title='No.of product sales')
```

```python
top_product = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(5)
top_product.plot(kind='bar',title='Top 5 product sales')
```

Region-wise Sales

```python
region_sales = df.groupby('Region')['Sales'].sum()
print(region_sales)
```

```python
region_sales.plot(kind='pie',title="Region-wise Sales")
```

Top Costomer

```python
top_customers = df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False).head(5)
top_customers.plot(kind='bar', title='Top Customers')
```
