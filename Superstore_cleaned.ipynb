```python
import pandas as pd
```

```python
df = pd.read_csv(r"/content/Raw_Superstore_Data.csv")
df.head(5)
```

```python
df.columns
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Sales'].fillna(df['Sales'].median(),inplace = True)
df['Quantity'].fillna(df['Quantity'].mean(),inplace = True)
df['Discount'].fillna(df['Discount'].mean(),inplace = True)
```

```python
df.head(7)
```

```python
df['Segment'].fillna(df['Segment'].mode())
```

```python
df.isnull().sum()
```

```python
categorial_cols = ['Customer Name','Segment','Region','City','Product Category','Product Name']
for cols in categorial_cols:
  df[cols].fillna(df[cols].mode()[0],inplace = True)
```

```python
categorical_null_counts = df[categorial_cols].isnull().sum()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates()
```

```python
df[df.duplicated()]
```

```python
df[df.duplicated(subset=['Order ID'])]
```

```python
df.drop_duplicates(inplace=True)
```

```python
print("After removal:", df.duplicated().sum())
```

```python
df['Month'] = [i%12+1 for i in range(len(df))]
df['Year'] = 2020

df['Date'] = pd.to_datetime(df['Year'].astype(str) + '-' + df['Month'].astype(str) + '-01')

df.drop(['Month', 'Year'], axis=1, inplace=True)
```

```python
df.to_csv("Superstore_Cleaned.csv", index=False)
```
